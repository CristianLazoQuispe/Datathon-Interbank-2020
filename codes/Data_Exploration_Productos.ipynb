# Load datasets

```python
productos = pd.read_csv('../resources/data/productos.csv')
productos = analyze.reduce_mem_usage(productos)

# Check uniques keys in table
print("rows in table :",productos.shape[0])
#print("key values    :",y_train.key_value.nunique())
productos.columns = ['description','Productos']
productos.head(2)
```

```python
productos
```

```python
def process_text1(text):
    if 'NO UTILIZADA' in text:
        return 1
    elif 'NO PREFERIDA' in text:
        return 2
    else:
        return 0
def process_text2(text):
    if 'CREDITO' in text:
        return 1
    else:
        return 0
productos['producto_class'] = productos.apply(lambda x: process_text1(x['description']), axis=1)
productos['producto_credito'] = productos.apply(lambda x: process_text2(x['description']), axis=1)
```
